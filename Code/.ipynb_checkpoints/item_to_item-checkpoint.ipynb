```python
def item_similarity(movieName): 
    try:
        user_inp = movieName
        inp = Tmovies_2[Tmovies_2['title'] == user_inp].index.tolist()
        inp = inp[0]
        movies_ratings['similarity'] = ratings_matrix_items.iloc[inp]
        movies_ratings.columns = ['movieId','title', 'genres','userId','rating','timestamp','similarity']
    except:
        print("No Recommendations")
```

```python
def recommendationBasedOnItemSimilarity(user_id):
    user_movie = movies_ratings[(movies_ratings.userId == user_id) & movies_ratings.rating.isin([5,4.5])][['title']]
    user_movie = user_movie.iloc[0,0]
    item_similarity(user_movie)
    movies_as_per_userChoice = movies_ratings.sort_values(["similarity"], ascending = False )
    movies_as_per_userChoice = movies_as_per_userChoice[movies_as_per_userChoice["similarity"] >= 0.45]['movieId']
    recommended_movies = list()
    df_recommended_item = pd.DataFrame()
    user2Movies = Tratings[Tratings['userId'] == user_id]['movieId']
    for movieId in movies_as_per_userChoice:
            if movieId not in user2Movies:
                df_new = Tratings[(Tratings.movieId == movieId)]
                df_recommended_item = pd.concat([df_recommended_item,df_new])
            best_recommendation = df_recommended_item.sort_values(["rating"], ascending = False )[1:10]
    return best_recommendation['movieId']
```

```python
def getMovieTitle(listMovieIDs):
    moviesIds = []
    [moviesIds.append(x) for x in listMovieIDs if x not in moviesIds]
    movie_titles= list()
    for id in moviesIds:
        movie_titles.append(Tmovies_2[Tmovies_2['movieId'] == id]['title'])
    return movie_titles
```
